# 이상치 검출

```python
data = pd.read_csv('mtcars.csv', index_col = [0])

def drop_outliers(data, n, features):
    outlier_indices = []
    for col in features:
        Q1 = np.percentile(data[col], 25)
        Q3 = np.percentile(data[col], 75)
        IQR = Q3 - Q1
        outlier_step = 1.5 * IQR
        outlier_list_col = data[(data[col] < Q1 - outlier_step) | (data[col] > Q3 + outlier_step)].index
        outlier_indices.extend(outlier_list_col)
    count_dic = {}
    for element in outlier_indices:
        if element not in count_dic:
            count_dic[element] = 0
        count_dic[element] += 1
    multiple_outliers = list(k for k, v in count_dic.items() if v > n)
    return multiple_outliers

outlier_to_drop = drop_outliers(data, 2, data.columns)
len(outlier_to_drop)
```

# PCA

```python
data = pd.read_csv('mtcars.csv', index_col = [0])

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_values = scaler.fit_transform(data)
data = pd.DataFrame(data_values, columns = data.columns, index = data.index)

from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pc = pca.fit_transform(data)
pc_data = pd.DataFrame(pc, columns = ['PC1', 'PC2'], index = data.index)

# 주성분별 분산 설명 확인
pca.explained_variance_ratio_
```

# log scaling

```python
data = pd.read_csv('mtcars.csv', index_col = [0])

for i in data.columns:
    print(i, 'skew:', data[i].skew())
    print(i, 'kurt:', data[i].kurt())
    print('----------------------------------')
    
data['carb'] = np.log1p(data['carb'])
print('After Log Scaling')
print('carb: ', data['carb'].skew(), data['carb'].kurt())
```
